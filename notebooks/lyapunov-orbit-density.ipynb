# Lyapunov sign and orbit density

This notebook is the slower companion to `reports/orbit-density-sidecar.md`.

The narrow question is the useful one: once a forcing word and a parameter pair are fixed, how does the orbit actually occupy `[0, 1]`?


## Why bother when we already have `λ`?

The Lyapunov exponent is still the right first summary because it tells us whether nearby trajectories contract or separate on average.

But sign alone does **not** tell us whether the tail collapses into a tiny support set, spreads across a few thin bands, or fills most of the interval.
That occupancy story is what the orbit-density layer adds.


```python
from lyaplab.orbit import summarize_orbit_density

cases = [
    ('AB stable', summarize_orbit_density('AB', 2.9, 3.1, burn_in=360, steps=1400, bins=24)),
    ('AB near-zero', summarize_orbit_density('AB', 2.9, 3.567, burn_in=360, steps=1400, bins=24)),
    ('AB chaotic', summarize_orbit_density('AB', 3.9, 3.95, burn_in=360, steps=1400, bins=24)),
]
[(label, round(summary.exponent, 4), summary.occupied_bins, round(summary.mean_x, 4)) for label, summary in cases]
```

That table is the core claim in compact form:

- the stable case collapses into a tiny support set
- the near-zero case already spreads across many more bins even before it looks fully chaotic
- the chaotic case fills most of the interval

So `λ < 0`, `λ ≈ 0`, and `λ > 0` are not just abstract signs. They come with visibly different occupancy geometry.


```python
more_cases = [
    ('AABAB stable', summarize_orbit_density('AABAB', 3.35, 3.35, burn_in=360, steps=1400, bins=24)),
    ('AABAB near-zero', summarize_orbit_density('AABAB', 2.9, 3.715, burn_in=360, steps=1400, bins=24)),
    ('AABAB chaotic', summarize_orbit_density('AABAB', 3.808, 3.808, burn_in=360, steps=1400, bins=24)),
]
[(label, round(summary.exponent, 4), summary.occupied_bins, round(summary.min_x, 4), round(summary.max_x, 4)) for label, summary in more_cases]
```

## What this does and does not prove

This is still a finite-time read of selected points, not a complete invariant-measure study over the whole plane.

But it already improves the public repo in one important way: the sequence planes no longer stop at sign fields.
We now have a concrete sidecar for reading what the orbit does after the sign has pointed us to an interesting location.
