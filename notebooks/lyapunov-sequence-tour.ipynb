# Lyapunov sequence tour

This notebook is the companion to `lyapunov-fractal-lab`: same core idea, slower pace.


```python
from lyaplab.core import lyapunov_exponent, classify_exponent
cases = [("AB", 2.9, 3.1), ("AB", 3.9, 3.95), ("AABAB", 3.4, 3.92)]
for seq, a, b in cases:
    value = lyapunov_exponent(seq, a, b)
    print(seq, a, b, round(value, 6), classify_exponent(value))
```

## Reading the sign

Negative means nearby trajectories contract on average. Positive means they separate on average.

The useful twist in this repo is that the forcing word over `A` and `B` changes the plane enough to be treated as part of the model.
